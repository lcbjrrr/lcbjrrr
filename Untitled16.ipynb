```python
import pandas as pd
dados = pd.read_csv('https://raw.githubusercontent.com/lcbjrrr/lcbjrrr/master/DistribuicaoAnestesicosBloqueadores%20-%20k.csv')
dados
```

```python
x= dados['QUANTIDADE']
maximo=x.max()
print('A maxima quantidade enviada foi de:', maximo)
y=dados['QUANTIDADE']
minimo=x.min()
print('A minima quantidade enviada foi de:',minimo)

z=dados['QUANTIDADE']
media=z.mean()
print('A media quantidade enviada foi de:',media)
```

Com o código acima, foi feita uma análise para descobrir o máximo, minimo e a média dos medicamentos enviados. Apos isso, pode-se perceber que a quantidade maxima enviada foi de 241650, já a minína 5, mostrando assim uma grande variação entre o máximo e minimo envio de medicamentos. Pode-se concluir que não é viavel  o ministério da saúde enviar mais de 250.000 mil medicamentos por dia, pois nos ultimos 2 anos esse numero ficou abaixo do valor, acarretando assim em um corte de gastos.

```python
soma_por_estado = dados.groupby('ESTADO').sum()['QUANTIDADE']
print(soma_por_estado)
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(40,10))
sns.barplot(x=soma_por_estado.index, y=soma_por_estado.values)
```

Com o uso do código acima, foi criado um gráfico de barras para visualizar a quantidade de medicamentos enviado para cada estados brasileiro. Feito isso, foi possível fazer uma análise, concluindo que os estados com mais medicamentos enviados foram os estados de São Paulo, Rio De Janeiro com uma diferença siginificativa para os demais estados. Isso já era o esperado pelo fato de os dois estados possuirem a maior população do pais, ressaltando também que o ministério da saúde deveria ter uma atenção especial para esses 2 estados, pois são de fato os estados que mais necessitam de remédios

```python
soma_por_data =dados.groupby('DATA DE ENTREGA').sum()['QUANTIDADE']
soma_por_data.head(30)
soma_viz = soma_por_data.reset_index()
soma_viz['ano_2020'] = soma_viz['DATA DE ENTREGA'].str.endswith("2020")
soma_viz
```

```python
viz1 = soma_viz.loc[soma_viz['ano_2020'] == True].copy()
viz1['DATA DE ENTREGA'] = pd.to_datetime(viz1['DATA DE ENTREGA'])
viz1 = viz1.sort_values(by='DATA DE ENTREGA')

viz2 = soma_viz.loc[soma_viz['ano_2020'] == False].copy()
viz2['DATA DE ENTREGA'] = pd.to_datetime(viz2['DATA DE ENTREGA'])
viz2 = viz2.sort_values(by='DATA DE ENTREGA')
```

```python
plt.figure(figsize=(26, 10))
plt.plot(viz1['DATA DE ENTREGA'], viz1['QUANTIDADE'], label='2020')
plt.plot(viz2['DATA DE ENTREGA'], viz2['QUANTIDADE'], label='2021')
plt.legend()
plt.show()
```
