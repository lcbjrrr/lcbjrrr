```python
import pandas as pd
nf = pd.read_csv('/content/drive/MyDrive/appro/Program - Página1.csv') 
nf
```

```python
import numpy as np
maximo=max(nf['vlrliquido'])
minimo=min(nf['vlrliquido'])
media=np.mean(nf['vlrliquido'])
print("Maximo:", maximo)
print("Minimo:", minimo)
print('Media', media)
```

Acima é possível averiguar o cálculo do Valor Líquido Máximo (armazenado em maximo), o de Valor Líquido Mínimo (armazenado em minimo) e a Média dos Valores Líquidos (armazenada em media). É possível analisar, através dos resultados obtidos, que a média dos valores é de 632,63 reais, o que contrasta grandemente com os valores máximo e mínimo, que mostram serem outliers. Ambos podem ser classificados como tal por estarem extremamente distantes da média: o valor máximo (de 53.000 reais), é 8377% maior que a média, enquanto o valor mínimo equivale à 0,002% da média. Com isso, podemos perceber que o maior gasto no período analisado foi de 53.000 reais, a média foi de 632,63 reais e o menor gasto foi de 0.18 reais, se tratando de dinheiro púlbico, podemos concluir que gastar 53.000 é um absurdo e um descaso com a população, ao invés desse dinheiro voltar em forma de benefícios.

```python
valorgastopartido = nf.groupby('sgpartido').sum()['vlrliquido']  
valorgastopartido
```

```python
valorgastopartido.plot(kind='pie')
```

Acima é possível visualizar um gráfico do estilo pizza contendo os dados dos valores gastos por partido, ou seja, um gráfico mostrando qual a contribuição de cada partido para os gastos totais. É possível perceber que, dentre os partidos que mais gastam, se destacam o PL, o PSD, o PROS e o PT, do maior para o menor (dentre os 4 que mais gastam). Já dentre os que menos contribuem para os gastos, se destacam o PDT, o PV, o PODE e o CIDADANIA, do menor para o maior (dentre os 4 que menos gastam).

```python
valorgastodeputado = nf.groupby('txnomeparlamentar').sum()['vlrliquido']  
valorgastodeputado
```

```python
valorgastodeputado.plot(kind='bar')
```

Acima é possível visualizar um gráfico de barras representando o Valor Gasto por Deputado, ou seja, o valor total gasto por cada deputado. Nele, é possível perceber que alguns deputados se destacam do resto, como o do deputado Giancobo, que se distancia em quase 33% (20 mil reais) do segundo colocado, Boca Aberta. Estes, junto com o deputado Roman (cujos valores se aproximamm de Boca aberta, perto da casa dos 60 mil reais), compõem os 3 maiores valores. Já os 3 menores valores se devem aos gastos dos deputados Paulo Eduardo Jardins, Gustavo Fruet e Felipe Francischini, respectivamente, que também se distanciam consideravelmente da média.

```python
valorgastoarea = nf.groupby('txtdescricao').sum()['vlrliquido']  
valorgastoarea
```

```python
valorgastoarea.plot(kind='bar')
```

Acima é possível visualizar um gráfico de barras indicando os Valores Gastos por Área, ou seja, por tipo de gasto. É possível perceber, através da análise do mesmo, que o tipo de gasto que mais contribui para o total de gastos é a Divulgação da Atividade Parlamentar, que corresponde a aproximadamente 30% dos gastos. Em segundo lugar, se concentra os gastos referentes à Manutenção de Escritório de Apoio à Atividade Parlamentar. Ademais, 3 tipos de gastos se destacam por contribuírem pouco no total, são eles, do menor para o maior: Serviço de Segurança Prestado por Empresa Especializada, Passagens Terrestres, Marítimas ou Fluviais e Fornecimento de Alimentação do Parlamentar.
