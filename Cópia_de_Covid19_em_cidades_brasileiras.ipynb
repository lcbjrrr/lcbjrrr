#Carrega o arquivo csv
Deve-se fazer upload aqui do arquivo "dataset_brazil_covid19_cities.csv"

```python
#Importa biblioteca Pandas
import pandas as pd
#Carrega o arquivo csv e já converte para um dataFrame
df = pd.read_csv('dataset_brazil_covid19_cities.csv')
#Imprime o dataFrame
df
```

#Sumarização: número de casos totais de covid19 por Estado
No número totais de casos da doença da convid19, percebe-se que o Estado com maior número de casos foi SP, com mais de 60 milhões de vítimas, sendo um dos principais fatores o número populacional ser o maior do país. Já RR e AC, um dos Estados menos populosos, têm os menores números de casos.

Percebe-se, assim, que o tamanho da poluação é, de certa forma, proporcional ao número de casos de covid19, pois terá mais indivíduos para se contaminar e além disso, tende a ter muitas pessoas concentradas em centros urbanos.

```python
#Agrupa todos estados com mesmos nomes
group1 = df.groupby('state')
#Selecion a columa cases do agrupamento, somas os valores para cada Estado, 
#e faz a ordenação do número de casos do maior para o menor e
# cria um nome para nova coluna
group1['cases'].sum().sort_values(ascending=False).reset_index(name ='cases')
```

#Sumarização: número de mortes totais cumulativas por covid19 por Estado
Como descrito anteriormente para números de casos de covid19, o número de mortes também tem influência no número populacional. Além disso existe mais um fator, que interfere no número de casos, e mais ainda no número de mortes por covid19, que é políticas públicas.

Na sumarização por número de casos, percebia-se que SP tinha mais do triplo do casos em relação a RJ. Mas quando se trata de número de mortes, tem menos do dobro. Isto pode ter haver com políticas públicas, como incentivos ao combate à pandemia, que reduziram casos mortais em SP, e talvez RJ não tenha seguido o mesmo caminho, tendo, assim, 7,6% de mortes do total de infectados, contra apenas 3,9% para SP.



```python
#Agrupa todos estados com mesmos nomes
group2 = df.groupby('state')
#Selecion a coluna deaths do agrupamento, somas os valores para cada Estado, 
#e faz a ordenação do número de mortes por covid19 do maior para o menor e
# cria um nome para nova coluna
group2['deaths'].sum().sort_values(ascending=False).reset_index(name ='deaths')
```

#Sumarização: número de casos totais cumulativos por covid19 por cidade
A cidade de São Paulo, igualmente ao Estado onde se encontra, principalmente por ser o maior centro urbano brasileiro e da América Latina, tem maior número de casos do país. Em segundo, se encontra Brasília, por ser o centro político brasileiro e muito populoso, como consequência.



```python
#Agrupa todas as cidades com mesmos nomes
group3 = df.groupby('name')
#Selecion a columa cases do agrupamento, somas os valores para cada Estado, 
#e faz a ordenação do número de casos do maior para o menor e
# cria um nome para nova coluna
group3['cases'].sum().sort_values(ascending=False).reset_index(name ='cases')
```

#Sumarização: número de mortes totais cumulativas por covid19 por cidade
A cidade de São Paulo continua na liderança, com proporção de 4,6 mortes para cada 100 infectados. Na cidade do Rio de Janeiro, com comportamento semelhante ao Estado onde se encontra, o número de letalidade é alto, com 11 mortes para cada 100 infectados, pelos possíveis motivos relacionados à políticas publicas, já mencionados.



```python
#Agrupa todas as cidades com mesmos nomes
group4 = df.groupby('name')
#Selecion a columa cases do agrupamento, somas os valores para cada Estado, 
#e faz a ordenação do número de casos do maior para o menor e
# cria um nome para nova coluna
group4['deaths'].sum().sort_values(ascending=False).reset_index(name ='deaths')
```

#Sumarização: quantidade de casos e mortes cumulativos de covid19 por mês
Nos dois próximos códigos, desconsiderando o mês de outubro de 2020, pelo fato dos dados cumulativos estarem mais baixos que os meses anteriores, provelmente porque quando foi criado o dataset alguns munícipios ainda não tinham liberados dados deste último mês.
Assim, considerando apenas os meses de julho, agosto e setembro, percebe-se que a quantidade de infectados e mortes diminuiram drasticamente com o passar dos meses. Além disso o percentual de mortes em relação ao número de infectados caiu. Este fator pode estar relacionado a novos métodos descobertos, de combate a covid19.

```python
#carrega o arquivo csv e convert para dataFrame
df2 = pd.read_csv('dataset_brazil_covid19_cities.csv')
#Renomeia a coluna data para month
df2 = df2.rename({'date': 'month'}, axis='columns')
#Convert dados do tipo string para datetime na coluna month
df2['month'] = pd.to_datetime(df2['month'])
#Faz sumarização com numero de casos totais de covid19 para cada, nomeando 
#a nova coluna de cases
df2.set_index('month',inplace=True)
df2['cases'].resample('m').sum().sort_values().reset_index(name ='cases')
```

```python
#Código similar ao código anterior, porém para número de mortes por convid19
df3 = pd.read_csv('dataset_brazil_covid19_cities.csv')
df3 = df3.rename({'date': 'month'}, axis='columns')
df3['month'] = pd.to_datetime(df3['month'])
df3.set_index('month',inplace=True)
df3['deaths'].resample('m').sum().sort_values().reset_index(name ='deaths')
```
