```
nts = pd.read_csv('https://raw.githubusercontent.com/Grupo-de-Pesquisa-em-Macro-Aplicada/Ibmec-Pesquisa-Macro-Aplicada/master/notas%20-%20ok.csv')
nts
```

```
nts[['AP1','AP2']]
```

```
import statsmodels.api as sm

ap1_ap2 = sm.add_constant(nts[['AP1','AP2']])
ap1_ap2
```

```
rl = sm.OLS(nts['AP3'], ap1_ap2).fit()
rl.summary()
```
